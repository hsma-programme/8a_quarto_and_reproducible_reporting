```python
# load the required libraries
import pandas as pd
```

```python
# load the dataset to a dataframe and display the first 5 rows
df = pd.read_csv('youtoxic_english_1000.csv')
df.head()
```

```python
# dataframe info
df.info()
```

```python
# save column names to a list
col_names = df.columns.tolist()
col_names = col_names[3:]

```

```python
# make a dictionary of the columns and their respective values
col_dict = {}
for col in col_names:
    col_dict[col] = df[col].value_counts().to_dict()

# print the dictionary one line at a time
for key, value in col_dict.items():
    print(key)
    print(value)
    print('\n')
```

```python
# plot the dictionary
import matplotlib.pyplot as plt

for key, value in col_dict.items():
    plt.bar(value.keys(), value.values())
    plt.xlabel(key)
    plt.ylabel('Count')
    plt.title(key)
    plt.show()
    
```
